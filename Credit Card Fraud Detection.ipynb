Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading the dataset to a Pandas DataFrame

```python
credit_card_data=pd.read_csv('creditcard.csv')
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

Distribution of legit transactions & fraudulent transactions

```python
credit_card_data['Class'].value_counts()
```

Separating the dataset for the analysis

```python
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Statistical measure of the Legit Transactions Amount Data

```python
legit.Amount.describe()
```

 Statistical measure of the Fraud Transactions Amount Data

```python
fraud.Amount.describe()
```

Compare the values for both the transactions

```python
credit_card_data.groupby('Class').mean()
```

Under-Sampling:
Build a Sample Dataset containing similar distribution of normal and fraudulent transactions

Creating a Sample Dataset of 492 transactions, as there are same number of fraudulent transactions

```python
legit_sample=legit.sample(n=492)
```

Concatenating two dataframes (legit_sample,fraud)

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

When axis = 0, the dataframe are concatenated row-wise,
if axis = 1, the dataframe are concatenated column-wise

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the Dataset into Features and Targets

Target = 'Class', Features = 'Other Columns of Dataframe'

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
X
```

```python
Y
```

Split the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y)
```

Given 20% of the data from the dataset to test and using to stratify to evenly distribute the data 

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Accuracy on the Training Data

```python
X_train_prediction=model.predict(X_train)
train_accuracy=accuracy_score(X_train_prediction,Y_train)
train_accuracy
```

Accuracy on Test Data

```python
X_test_prediction=model.predict(X_test)
test_accuracy=accuracy_score(X_test_prediction,Y_test)
test_accuracy
```

Since the Accuracy Scores of the training data and test data are almost similar so, it can be said that the model is a perfect fit.
